```python
#Fazendo o calculo de Assimetria e Curtose
import pandas as pd
from scipy.stats import skew, kurtosis

#Carregar o arquivo Vendas para um DataFrame
tabela = pd.read_excel("dvendas.xlsx")

#Calcular a assimetria e a curtose da coluna "Qtde_Vendida"
assimetria = skew(tabela["Qtd_Vendida"])
curtose = kurtosis(tabela["Qtd_Vendida"])

#Mostrar Assimetria e Curtose
print(f'Assimetria: {assimetria}')
print(f'Curtose: {curtose}')
```
