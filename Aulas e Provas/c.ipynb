```python
import numpy as np
from scipy import stats

def calculate_statistics_and_probability(data, value):
    # Calcula a média e o desvio padrão
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # ddof=1 para usar o desvio padrão amostral

    # Calcula a probabilidade de um valor específico na distribuição normal
    z_score = (value - mean) / std_dev
    probability = stats.norm.cdf(z_score)

    return mean, std_dev, probability

def main():
    # Dados de exemplo
    data = [17, 17, 17, 20, 23, 25, 22, 17, 14, 22, 23,
16,
19,
21,
22,
25,
25,
25,
25,
21,
17,
17,
13,
13,
13,
13,
14,
15,
15,
15,
18,
20,
19,
20,
21,
22,
23,
24,
25,
25,
26,
27,
28,
27,
30,
25,
22,
21,
20,
22,
26,
27,
28,
29,
30,
25,
25,
25,
20,
29,
30,
25,
30,
28,
30,
23,
22,
26,
30,
27,
26,
27,
28,
29,
30,
21,
20,
27,
30,
28,
]
    value = 21  # Valor específico para calcular a probabilidade

    mean, std_dev, probability = calculate_statistics_and_probability(data, value)

    print(f"Média: {mean}")
    print(f"Desvio Padrão: {std_dev}")
    print(f"Probabilidade de um valor ser menor ou igual a {value}: {probability * 100:.2f}%")
    print(f"Probabilidade de um valor ser maior que {value}: {(1 - probability) * 100:.2f}%")

if __name__ == "__main__":
    main()
```
