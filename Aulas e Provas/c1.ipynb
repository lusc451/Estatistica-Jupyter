```python
import numpy as np
from scipy import stats

def calculate_statistics_and_probability(data, a, b):
    # Calcula a média e o desvio padrão
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # ddof=1 para usar o desvio padrão amostral

    # Calcula os valores z para os limites a e b
    z_a = (a - mean) / std_dev
    z_b = (b - mean) / std_dev

    # Calcula a probabilidade acumulada até os valores z
    prob_a = stats.norm.cdf(z_a)
    prob_b = stats.norm.cdf(z_b)

    # A probabilidade de estar entre a e b é a diferença entre as duas probabilidades
    probability_between = prob_b - prob_a

    return mean, std_dev, probability_between

def main():
    # Dados de exemplo
    data = [17, 17, 17, 20, 23, 25, 22, 17, 14, 22, 23,
16,
19,
21,
22,
25,
25,
25,
25,
21,
17,
17,
13,
13,
13,
13,
14,
15,
15,
15,
18,
20,
19,
20,
21,
22,
23,
24,
25,
25,
26,
27,
28,
27,
30,
25,
22,
21,
20,
22,
26,
27,
28,
29,
30,
25,
25,
25,
20,
29,
30,
25,
30,
28,
30,
23,
22,
26,
30,
27,
26,
27,
28,
29,
30,
21,
20,
27,
30,
28,
]
    a = 20  # Limite inferior
    b = 21  # Limite superior

    mean, std_dev, probability_between = calculate_statistics_and_probability(data, a, b)

    print(f"Média: {mean}")
    print(f"Desvio Padrão: {std_dev}")
    print(f"Probabilidade de um valor estar entre {a} e {b}: {probability_between * 100:.2f}%")

if __name__ == "__main__":
    main()
```
